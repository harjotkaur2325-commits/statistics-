 Question 1: What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
Answer- Type | (False positive) – It occurs when you reject the null hypothesis(h0) when it is actually true. Example – a new drug is tested for effectiveness. H0 – Drug has no effect.   (if the test says the drug is working but in reality, it doesn’t)
Impact on decision- making - you may take unnecessary action, e.g., approve an ineffective drug, leading to wasting resources.
Type || Error (False negative) - It occurs when you fail to reject the null hypothesis (h0) when it is actually false. Example- Using the same drug. Ho-Drug has no effect. (If the test says the drug doesn’t work but in reality, it does)
Impact on decision- making – you may miss opportunity, e.g., reject a beneficial treatment or fail to act on important findings.

Question 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
Answer – The p- Value is the probability of observing the sample data assuming the null hypothesis (h0) is true. It measures how compatible your data is with the null hypothesis.
Interpret the p- value:-
P-value	Decision	Meaning
p ≤ 0.05	Reject H₀	Evidence against null → significant result
p > 0.05	Fail to reject H₀	Insufficient evidence → not significant
Example-      H0: A new teaching method does not affect student scores.
Test gives p-value = 0.02, interpretation: - p = 0.02 < 0.05  Reject H0

Question 3: Explain the difference between a Z-test and a T-test, including when to use each.
Answer – Z-test :- A Z-test is used to determine whether there is significant difference between sample and population means when the population standard deviation is known.
Key Points – 1. Requires large sample size (n >=30) for normal approximation.
2. Uses Z-distribution (standard normal distribution).
3. Commonly used for population mean.
Example:
•	A factory claims the average weight of a product is 50g, population SD = 5g, sample n = 40. Test if the sample differs.
•	Use Z-test because population standard deviation is known.
T-test - A T-test is used when the population standard is unknown and the sample size is small (n < 30). It uses the t-distribution, which adjusts for small sample sizes.
Key Points:
1.	Accounts for extra uncertainty from estimating sigma with sample standard deviation (s).
2.	Common types:
1.	One-sample T-test → compare sample mean with population mean
2.	Independent two-sample T-test → compare means of two independent groups
3.	Paired T-test → compare means of related samples (e.g., before-after measurements)
Example: Test whether students’ average exam score differs from expected 75, sample size = 20, , sigma unknown → Use T-test.

Question 4: What is a confidence interval, and how does the margin of error influence its width and interpretation?
Answer – A confidence interval is range of values used to estimate an unknown population parameter (like the population mean) based on sample data. It provides an interval that is likely to contain the true population value with a certain level of confidence. Example
Suppose you survey some students and find the average study time = 5 hours.
Your 95% confidence interval might be: 4.5 hours to 5.5 hours
This means:  We are 95% confident that the true average study time of all students is between 4.5 and 5.5 hours.
Margin of error – The margin of error is how much the estimate can go up or down. Let’s see how the margin of error changes the Width: -
Large margin of error – Wide interval
Example: Average = 60, margin of error = 10, confidence interval = 70 to 80
Interpretation: we only know the true value is somewhere in a big range, so the estimate is less accurate.
Small margin of error – narrow interval
Example = Average = 50, margin of error = 2, confidence interval = 48 to 52
Interpretation: The true value is likely very close to 52, so the estimate is more reliable.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
Answer – Purpose of ANOVA – The Analysis of Variance(ANOVA) is used to test whether the means of three or more groups are significantly different from each other. Instead of performing many separate tests, ANOVA check all groups at once.
Example – suppose a researcher wants to compare student scores under the three teaching methods:
1.	Method A       2.     Method B     3. Method C
ANOVA tests whether all these groups average are the same or if at least one is different.
It does this using the F-test, which compares between-group variance and within-group variance.
Assumptions of ANOVA – for ANOVA results to be reliable, these assumptions must be satisfied:
1.	Normality: The data in each group should follow a normal distribution.
2.	Independence of observations: Each observation should be independent of others.
3.	Homogeneity of variance: All groups should have similar variances.
How ANOVA Extends hypothesis Testing: Normally, tests like the t-test can only compare two groups at a time. ANOVA extends this by allowing us to compare three or more groups means in a single test. This helps reduce multiple testing errors and provides a single overall conclusion.















```python
#Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
import numpy as np
from scipy import stats

# sample data
data = np.array([52, 48, 51, 49, 50, 53, 47, 52, 50, 51])

#population parameters
mu = 50      #population mean
sigma = 5    #population standard deviation
n = len(data)


# sample mean
x_bar = np.mean(data)

# Z- test formula
Z = (x_bar - mu)/(sigma/np.sqrt(n))

# calculate p-value (two - tailes test)
p_value = 2 * (1 - stats.norm.cdf(np.abs(Z)))

print("sample means:", x_bar)
print("Z-score:", Z)
print("p-value:", p_value)

# decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")


```

```python
#Question 8: Generate multiple samples from a non-normal distribution and implement
# the Central Limit Theorem using Python

import numpy as np
import matplotlib.pyplot as plt

# generate data from a non- normal distribution(exponential)
population = np.random.exponential(scale = 2, size= 10000)

# take many samples and compute their means
sample_size = 30
num_sample = 1000
sample_means = []

for i in range(num_sample):
    sample = np.random.choice(population, size = sample_size)
    sample_means.append(np.mean(sample))

# plot the histogram of sample means
plt.hist(sample_means, bins = 30, edgecolor = 'black')
plt.title("Central limit theorem")
plt.xlabel("sample mean")
plt.ylabel("frequency")
plt.show()
```

```python
from matplotlib.lines import lineStyles
#Question 9: Write a Python function to calculate and visualize the confidence interval
# for a sample mean.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval(data, confidence = 0.95):
  n = len(data)
  mean = np.mean(data)
  std_error = stats.sem(data) # standard error

  # t critical value
  t_value = stats.t.ppf((1+confidence)/2, df = n-1)

  margin_error = t_value * std_error
  lower =  mean - margin_error
  upper = mean + margin_error

  print("sample mean:", mean)
  print(f"{confidence*100}% confidence interval:", (lower, upper))

  # visualization
  plt.hist(data, bins = 10, edgecolor = 'black', alpha = 0.7)
  plt.axvline(mean, color = "red", label = "Sample mean")
  plt.axvline(lower, color = "green", linestyle= "--", label = "lower CI")
  plt.axvline(upper, color = "yellow", linestyle= "--", label = "upper CI")

  plt.title("Confidence interval")
  plt.legend()
  plt.show()

# example dataset

Data = [12, 15, 14, 10, 18, 16, 14, 13, 17, 19]

confidence_interval(Data)

```

```python
##Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
## observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

#oberved frequencies
observed = np.array([8, 10, 9, 11, 12, 10])

#expected frequencies
expected = np.array([10, 10, 10, 10, 10, 10])

# chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("chi-square statistic:", chi_stat)
print("p-value:", p_value)

#decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

```

interpretation:
* if:  p-value < 0.05  
--> Reject the null hypothesis

--> observed distribution does not match expected distribution

* p-value> 0.05

--> Fail to reject the null hypothesis

--> observed distribution matches expected distribution
